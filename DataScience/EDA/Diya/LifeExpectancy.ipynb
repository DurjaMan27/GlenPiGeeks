```python
import pandas as pd
import matplotlib.pyplot as plt
lf_df = pd.read_csv('/home/ec2-user/SageMaker/GlenPiGeeks/DataScience/Data/Diya/LifeExpectancy.csv', thousands=',', header = 0)
lf_df.drop(lf_df.index[0], axis=0, inplace=True)
lf_df.drop(lf_df.index[0], axis=0, inplace=True)
plt.figure(figsize=(20,12))
plt.plot('State/County',  'Male life expectancy, 1985 (years)', data=lf_df, markerfacecolor='green', markersize=12, color='skyblue', linewidth=4)
plt.plot('State/County',  'Female life expectancy, 1985 (years)', data=lf_df, marker='X', color='olive', linewidth=2)
plt.xticks(rotation=90)
plt.xlabel  = 'this is a xlabel \n  with newlines!'
plt.ylabel  = 'this is vertical\ntest'
plt.legend()
plt.show()
```
