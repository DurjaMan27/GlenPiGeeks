```python
import pandas as pd
import matplotlib.pyplot as plt
pc_df = pd.read_csv('/home/ec2-user/SageMaker/GlenPiGeeks/DataScience/Data/Katie/prison_custody_by_state.csv', thousands=',', header = 0)
pc_df.drop(pc_df.columns[1], axis=1, inplace=True)
plt.figure(figsize=(20,12))
plt.plot('2001',  'Federal', data=pc_df, markerfacecolor='green', markersize=12, color='skyblue', linewidth=4)
plt.plot('2002',  'Alabama', data=pc_df, marker='X', color='olive', linewidth=2)
plt.plot('2003',  'Alaska', data=pc_df, marker='^', color='red', linewidth=2, linestyle='dashed')
plt.xticks(rotation=90)
plt.xlabel  = 'this is a xlabel \n  with newlines!'
plt.ylabel  = 'this is vertical\ntest'
plt.legend()
plt.show()
```
